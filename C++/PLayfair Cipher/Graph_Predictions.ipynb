# Graph Visualization and Result Analysis for PlayFair Cipher

```python
import matplotlib.pyplot as plt
```

```python
from csv import reader

x = []    # no. of words
y = []    # Time in ms


with open('sizeVariation_time_Scratch.txt', 'r') as file:
    data = reader(file, delimiter=' ')
    transposed = (list(item) for item in zip(*data))
    

y,x=transposed
```

```python
y = list(map(int, y))
```

```python
plt.plot(x,y,"o:")
plt.grid(True)
plt.xlabel('No. of Words')
plt.ylabel('Time taken (in ms)')

plt.title('Performance graph Serial -: Variable Number of Words')
plt.xticks(rotation=45)
for i in range(len(x)):
    if(i==8):
        plt.text(6,y[i]-450,"       "+str(y[i])+'ms')
        break
    else:
        s="  "
    s+=str(y[i])
    #s+=')'
    
    plt.text(x[i],y[i],s)
plt.show()
```

```python
from csv import reader

words = []    # no. of words
time_2 = []    # Time


with open('Sample_Task_sizeVariation_time_Scratch.txt', 'r') as file:
    data = reader(file, delimiter=' ')
    transposed = (list(item) for item in zip(*data))
   

time_2,words=transposed
```

```python
time_2 = list(map(int,time_2))
```

```python
plt.grid(True)
plt.plot(words,time_2,"ro:")
plt.xlabel('No. of Words')
plt.ylabel('Time taken (in ms)')

plt.title('Performance Graph Task Parallelism+Data Parallelism :- Words Variation')


plt.xticks(rotation=45)

for i in range(len(words)):
    if(i==8):
        plt.text(6,time_2[i],"        "+str(time_2[i])+'ms')
        break
    s="   "
    s+=str(time_2[i])
    #s+=')'
    plt.text(words[i],time_2[i],s)

plt.show()
```

```python
# fig = plt.figure(figsize=(6,15))
plt.grid(True)

# scale_factor = 1000
# plt.ylim(0* scale_factor, 16000)
plt.plot(words,time_2,"ro:")
plt.plot(x,y,"o:")



plt.xlabel('No. of Words')
plt.ylabel('Time taken (in ms)')

plt.title('Comparison b/w Task Parallelism+Data Parallelism and Serial :- Words Variation')
plt.xticks(rotation=45)

for i in range(len(words)):
    if(i==8):
#         plt.text(words[i],time_2[i],str(time_2[i]))
        plt.text(6,y[i],"      "+str(y[i])+'ms')
        break
    elif(i<=4):
        continue
    s="  "
    s1="  "
    s+=str(time_2[i])
    s1+=str(y[i])
    #s+=')'
    plt.text(words[i],time_2[i],s)
    plt.text(words[i],y[i],s1)

plt.show()
```

## Task Parallelism + Data Parallelism 1.6x faster from Serial (Encryption/Decryption)

```python
serial_y = y[0:5]
new_x = x[0:5]
parallel_y = time_2[0:5]

plt.grid(True)

# scale_factor = 1000
# plt.ylim(0* scale_factor, 16000)
plt.plot(new_x,parallel_y,"ro:")
plt.plot(new_x,serial_y,"o:")



plt.xlabel('No. of Words')
plt.ylabel('Time taken (in ms)')

plt.title('Comparison b/w Task Parallelism+Data Parallelism and Serial :- Words Variation < 60K')
plt.xticks(rotation=45)

for i in range(len(new_x)):
    if(i==4):
        plt.text(3,serial_y[i],"           "+str(serial_y[i])+"ms")
        plt.text(3,parallel_y[i],"            "+str(parallel_y[i])+"ms")
        break
    elif(i==0):
        continue
    s="  "
    s1="  "
    s+=str(serial_y[i])
    s1+=str(parallel_y[i])
    #s+=')'
    plt.text(new_x[i],serial_y[i],s+"ms")
    plt.text(new_x[i],parallel_y[i],s1+"ms")

plt.show()
```
